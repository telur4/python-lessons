```python
# HTTP経由でコンテンツを取得する
import requests

res = requests.request('get', 'https://codezine.jp/')
print(res.text)
```

```python
# HTTP POSTによる通信
import requests

res = requests.post('https://wings.msn.to/tmp/post.php', data = {'name': '佐々木新之助'})
print(res.text)
```

```python
# JSONデータを取得する
import requests

res = requests.get('https://wings.msn.to/tmp/books.json')
bs = res.json()
print(bs['books'][0]['title'])
```
