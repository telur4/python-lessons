```python
# 円グラフを作成する
import matplotlib.pyplot as plt

%matplotlib inline

data = [60, 20, 10, 5, 3, 2]

# dataを円グラフとして可視化
plt.pie(data)

# 円グラフを楕円から真円する
plt.axis('equal')

plt.show()
```

```python
# 円グラフにラベルを設定する
import matplotlib.pyplot as plt

%matplotlib inline

data = [60, 20, 10, 5, 3, 2]
labels = ['Apple', 'Orange', 'Banana', 'Pineapple', 'Kiwifruit', 'Strawberry']

# dataにlabelsのラベルをつけ、円グラフで可視化
plt.pie(data, labels=labels)

plt.axis('equal')

plt.show()
```

```python
# 特定の要素を目立たせる
import matplotlib.pyplot as plt

%matplotlib inline

data = [60, 20, 10, 5, 3, 2]
labels = ['Apple', 'Orange', 'Banana', 'Pineapple', 'Kiwifruit', 'Strawberry']
explode = [0, 0, 0.1, 0, 0, 0]

# dataにlabelsのラベルをつけ、Bananaを目立たせた円グラフで可視化
plt.pie(data, labels=labels, explode=explode)

plt.axis('equal')

plt.show()
```
