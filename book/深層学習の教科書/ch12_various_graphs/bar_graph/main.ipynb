```python
# 棒グラフを作成する
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y = [12, 41, 32, 36, 21, 17]

# 棒グラフを作成
plt.bar(x, y)

plt.show()
```

```python
# 横軸にラベルを設定
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y = [12, 41, 32, 36, 21, 17]

labels = ['Apple', 'Orange', 'Banana', 'Pineapple', 'Kiwifruit', 'Strawberry']

# 棒グラフを作成し横軸にラベルを設定
plt.bar(x, y, tick_label=labels)

plt.show()
```

```python
# 積み上げ棒グラフを作成する
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]

labels = ['Apple', 'Orange', 'Banana', 'Pineapple', 'Kiwifruit', 'Strawberry']

# 積み上げ棒グラフを作成し横軸にラベルを設定
plt.bar(x, y1, tick_label=labels)
# bottomに下の棒グラフを設定
plt.bar(x, y2, bottom=y1)

# 系統ラベルを設定
plt.legend(('y1', 'y2'))
plt.show()
```
