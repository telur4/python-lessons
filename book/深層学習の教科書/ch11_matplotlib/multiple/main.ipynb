```python
# 1つのグラフに2種類のデータをプロットする
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)      # １つ目のデータ
y2 = np.cos(x)      # 2つ目のデータ

plt.title('graphs of trigonometric functions')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)

positions = [np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2]
labels = ['90', '180', '270', '360']
plt.xticks(positions, labels)

# データx, y1をグラフにプロットし黒で表示
plt.plot(x, y1, color='k')

# データx, y2をグラフにプロットし青で表示
plt.plot(x, y2, color='b')

plt.show()
```

```python
# 系列ラベルを設定する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('graphs of trigonometric functions')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True)

positions = [np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2]
labels = ['90', '180', '270', '360']
plt.xticks(positions, labels)

# データx, y1をグラフにプロットしラベルをつけて黒で表示
plt.plot(x, y1, color='k', label='y=sin(x)')

# データx, y2をグラフにプロットしラベルをつけて青で表示
plt.plot(x, y2, color='b', label='y=cos(x)')

# 系列ラベルを設定
plt.legend(['y=sin(x)', 'y=cos(x)'])

plt.show()
```

```python
# 図の大きさを設定する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

# 図の大きさを設定する
# plt.figure(figsize=(横の大きさ, 縦の大きさ)) 単位はインチ
plt.figure(figsize=(4, 4))

plt.plot(x, y)
plt.show()
```

```python
# サブプロットを作成する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

# Figureオブジェクトを作成
fig = plt.figure(figsize=(9, 6))
print(fig)

# 2x3のレイアウトの上から2行目、左から2列目にサブプロットオブジェクトを作成
ax = fig.add_subplot(2, 3, 5)
print(ax)

# データx, yをプロットし表示
ax.plot(x, y)

# グラフがどこに追加されるか確認するため空白部分をサブプロットで埋める
axi = []
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i + 1)

plt.show()
```

```python
# サブプロットの周りの余白を調節する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

# 図内のサブプロット間を、縦横ともに1の割合で開ける
# サブプロットのサイズを1とする
plt.subplots_adjust(wspace=1, hspace=1)

ax.plot(x, y)

axi = []
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i + 1)

plt.show()
```

```python
# サブプロット内のグラフの表示範囲を設定する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

plt.subplots_adjust(wspace=1, hspace=1)

# サブプロットaxのグラフのy軸の表示範囲を[0, 1]を設定する
ax.set_ylim([0, 1])

ax.plot(x, y)

axi = []
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i + 1)

plt.show()
```

```python
# サブプロット内のグラフの要素に名前を設定する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

plt.subplots_adjust(wspace=1, hspace=1)

# サブプロットaxのグラフのタイトルを設定
ax.set_title('y=sin(x)')

# サブプロットaxのグラフのx軸、y軸に名前を設定
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.plot(x, y)

axi = []
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i + 1)

plt.show()
```

```python
# サブプロット内のグラフの軸にメモリを設定する
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

positions = [0, np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2]
labels = ['0', '90', '180', '270', '360']

fig = plt.figure(figsize=(9, 6))

ax = fig.add_subplot(2, 3, 5)

plt.subplots_adjust(wspace=1, hspace=1)

ax.grid(True)

ax.set_title('y=sin(x)')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

# サブプロットaxのグラフのx軸にメモリを設定
ax.set_xticks(positions)
ax.set_xticklabels(labels)

ax.plot(x, y)

axi = []
for i in range(6):
    if i == 4:
        continue
    fig.add_subplot(2, 3, i + 1)

plt.show()
```
